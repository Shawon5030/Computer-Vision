
#  cv2.imread()

```python
# Load color image
color_image = cv2.imread('image.jpg', cv2.IMREAD_COLOR)
# Load grayscale image
grayscale_image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
# Load image with alpha channel
unchanged_image = cv2.imread('image.png',cv2.IMREAD_UNCHANGED)

#Image read :
    # if we wanna show a image with the help of cv2 first of all we need to read the image
    # for reading a image we need to use cv2 function imread()
    
#imread function: 
    #imread have 2 parameter first one is image location , anothter is flag 
    
#flag:
   #cv2.IMREAD_COLOR: This flag is represented by the integer value 1. It loads a color image.
   #cv2.IMREAD_GRAYSCALE: This flag is represented by the integer value 0. It loads the image in grayscale mode.
   #cv2.IMREAD_UNCHANGED: This flag is represented by the integer value -1. It loads the image as is, including the alpha channel if it exists.

# Display the image in a window named 'color_image','grayscale_image','unchanged_image'

cv2.imshow('color_image' ,     color_image)
cv2.imshow('grayscale_image' , grayscale_image )
cv2.imshow('unchanged_image' , unchanged_image )

# Wait for a key press and close the window when any key is pressed
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# cv2.resize()

```python
#let's first read a image
img1 = cv2.imread('cat.jpg', cv2.IMREAD_COLOR)
resize = cv2.resize(img1,(380,200))#width ,height

#why we use resize funciton?
   #cv2.resize() function commonly used for resizing images is cv2.resize().
    
#Who many parameters  cv2.resize() function have?
  #1  src: The source image.
  #2. (dsize): Output image size, specified as a tuple (width, height).
    # (this 2 must be in)
    
# Display the image in a window named 'Image'
cv2.imshow('Cat image' , resize )

# Wait for a key press and close the window when any key is pressed
cv2.waitKey(0)
cv2.destroyAllWindows()
```
